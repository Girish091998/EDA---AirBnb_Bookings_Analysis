## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

airbnb_df.drop_duplicates(inplace=True)                                              # This removes duplicate rows if any, which in turn eliminates data duplicacy.

airbnb_df.fillna({'reviews_per_month':0}, inplace=True)                              # We will fill all NaN with 0 as it doesn't have any review in few of the months

airbnb_df["name"].fillna("unknown", inplace=True)                                    # We will replace NaN with "unknown"

airbnb_df["host_name"].fillna("unknown", inplace=True)                               # We will replace NaN with "unknown"

edit = airbnb_df["price"].mean()                                                     # Will calculate the mean price

airbnb_df.replace(0, edit, inplace=True)                                             # Will then replace price 0 with mean value

airbnb_df.isna().sum()                                                               # We have data["reviews_per_month"] column with 0 null values
````

#### 2. **What are the seasonal, monthly, and weekly booking trends?**

````python
# Assuming you have a date column named 'last_review' in the format 'yyyy-mm-dd'
airbnb_df['last_review'] = pd.to_datetime(airbnb_df['last_review'])

# Extract seasonal, monthly, and weekly trends
airbnb_df['season'] = airbnb_df['last_review'].dt.quarter
airbnb_df['month'] = airbnb_df['last_review'].dt.month
airbnb_df['week'] = airbnb_df['last_review'].dt.isocalendar().week

# Plot seasonal trend
plt.figure(figsize=(8, 6))
sns.countplot(x='season', data=airbnb_df)
plt.xlabel('Season')
plt.ylabel('Bookings Count')
plt.title('Seasonal Booking Trends')
plt.show()

# Plot monthly trend
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=airbnb_df)
plt.xlabel('Month')
plt.ylabel('Bookings Count')
plt.title('Monthly Booking Trends')
plt.show()

# Plot weekly trend
airbnb_df.dropna(subset=['week'], inplace=True)  # Remove rows with missing 'week'
airbnb_df['week'].fillna(-1, inplace=True)  # Fill missing 'week' with -1

plt.figure(figsize=(12, 6))
sns.countplot(x='week', data=airbnb_df)
plt.xlabel('Week')
plt.ylabel('Bookings Count')
plt.title('Weekly Booking Trends')
plt.show()
````

#### 3. **Which locations are the most popular destinations?**

````python
# Plot Most Popular Destinations
plt.figure(figsize=(10, 6))
popular_destinations = airbnb_df['neighbourhood'].value_counts().head(10)
sns.barplot(x=popular_destinations.index, y=popular_destinations.values)
plt.xlabel('neighbourhood')
plt.ylabel('Number of Bookings')
plt.title('Top 10 Most Popular Destinations')
plt.xticks(rotation=45)
plt.show()
````

####4. How does the availability of listings vary across months?

````python
# Extract month from the 'availability 365' column
airbnb_df['month'] = airbnb_df['availability_365'].apply(lambda x: (x // 30) + 1)

# Plot Availability of Listings across Months
plt.figure(figsize=(10, 6))
availability_counts = airbnb_df['month'].value_counts().sort_index()
sns.barplot(x=availability_counts.index, y=availability_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of Listings')
plt.title('Availability of Listings across Months')
plt.show()
````
